#### Task - 1 (Where Name is missing fill the value in place by extracting the name from the Email-id column.)

```python
# extract names from email address column
df['Name'] = df['Name'].fillna(df['Email'].str.extract(r'(^[\w.-]+)@', expand=False))

# capitalize first letter of each name
df['Name'] = df['Name'].str.title()
```

#### Task - 5 (In the Department column make the data proper with the useful functions and rename the wrong spelled words in the department column.)



                                                                            


```python
# convert all department names to  remove leading/trailing whitespaces
df1['Department'] = df1['Department'].str.strip()

# create a dictionary of misspelled words and their correct spellings
misspelled_words = {'SLES': 'Sales',
                    'SALES': 'Sales',
                    'DaTa SciENce': 'Data Science',
                    'hR': 'Hr',
                    'HR':'Hr',
                     'IT DeVlopemENT' : 'IT Development',
                     'OperarYION':'operation'}

# replace misspelled words with their correct spellings
df1['Department'] = df1['Department'].replace(misspelled_words)
```

#### Task - 7 (Clean and Format the Phone number column and it should show like (91-xxxxxxxx))

```python
df1['Phone'].describe()
```

```python
miss_phone = np.where(df1['Phone'].isnull()==True)
```

```python
miss_phone
```

```python
len(miss_phone[0])
```

```python
# clean and format the phone number column

df1['Phone'] = df1['Phone'].astype(str)   # convert to string data type
df1['Phone'] = df1['Phone'].str.replace('[^0-9]', '')   # remove all non-numeric characters
df1['Phone'] = df1['Phone'].str[-10:]   # keep only the last 10 digits
df1['Phone'] = '91-' + df1['Phone']     # add the country code

```

#### Task - 8 (Highlight the Top 10 Employees with highest salary in the dataset.)

```python
# sort the DataFrame by the 'Salary' column in descending order and select the top 10 employees by index

top_10_employees_index = df1['Salary'].sort_values(ascending=False).head(10).index

# create a new DataFrame with the same data and highlight the top 10 employees with a background color
df2 = df1.style.apply(lambda x: ['background-color: #ffcccc' if x.name in top_10_employees_index else '' for i in x], axis=1)
```
